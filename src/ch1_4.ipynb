```python
def palindrome_permutation(string):
  if len(string)==0 or len(string)==1:
    return True
  else:
    counts_charset = [0 for i in range(128)]
    for char in string:
      if char.lower() != " ":
        counts_charset[ord(char.lower())]+=1
  odd_counts=0
  for element in counts_charset:
    if element%2!=0:
      odd_counts+=1
  return ( odd_counts<=1)
```

```python
assert(palindrome_permutation("Tact Coa")== True)
assert(palindrome_permutation("Abiola")== False)
assert(palindrome_permutation("")== True)
assert(palindrome_permutation("b")== True)
assert(palindrome_permutation("aa")== True)
assert(palindrome_permutation("RERVE")== True)
assert(palindrome_permutation("baba")== True)
assert(palindrome_permutation("facebook")== False)
```
