```python
def string_compression(string):
  if len(string)<=2:
    return string
  compression_list = []
  string_length = len(string)
  current_char = string[0]
  current_char_counts = 0
  iteration_counter = 0
  for char in string:
    iteration_counter+=1
    to_check = char
    if to_check == current_char:
      current_char_counts+=1
    else:
      if len(compression_list)+2 >= string_length:
        return string
      else:
        compression_list.append(current_char)
        compression_list.append(str(current_char_counts))
        current_char = to_check
        current_char_counts = 1
    if iteration_counter==string_length:
        compression_list.append(current_char)
        compression_list.append(str(current_char_counts))
  print(compression_list)
  return "".join(compression_list)
```

```python
assert(string_compression("")=="")
assert(string_compression("a")=="a")
assert(string_compression("ab")=="ab")
assert(string_compression("aa")=="aa")
assert(string_compression("aabcccccaaa")=="a2b1c5a3")
```
