```python
#simple solution
def urlify(string):
  return "%20".join(string.split(" "))
```

```python
assert(urlify(" ")=="%20")
assert(urlify("I am.")=="I%20am.")
```

```python
#without any builtin finction
def urlify_2(string):
  string_to_return =[]
  for char in string:
    if char==" ":
      string_to_return.append("%20")
    else:
      string_to_return.append(char)
  return "".join(string_to_return)
```

```python
assert(urlify_2(" ")=="%20")
assert(urlify_2("I am.")=="I%20am.")
```
