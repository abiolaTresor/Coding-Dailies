```python
def rotate_matrix(np_array):
  nrows = np_array.shape[0]
  ncols = np_array.shape[1]
  new_arr = np.array([[0]*nrows]*ncols)
  print(new_arr.shape)
  for j in range( ncols):
    for i in range(nrows):
      new_arr[ncols-j-1, i] = np_array[i,j]
  return new_arr
```
