```python
def is_unique(string):
  cache = dict()
  for char in string:
    if char in cache:
      return False
    else:
      cache[char]=1
  return True
```

```python
assert(is_unique("abiola")==False)
```

```python
assert(is_unique("abcdef")==True)
```

```python
def is_unique_2(string):
  for index in range(len(string)):
    current_char = string[index]
    for other_index in range(len(string)):
      if other_index != index:
        other_current_char = string[other_index]
        if other_current_char == current_char:
          return False
  return True
      
```

```python
assert(is_unique_2("abiola")==False)
```

```python
assert(is_unique_2("abcdef")==True)
```

```python
 ##with hypothesis on the encoding (128--ascii)
 import string
 def is_unique_3(string):
   bool_list = [False]*128
   for char in string:
     if bool_list[ord(char)]:
       return False
     bool_list[ord(char)] = True
   return True
```

```python
assert(is_unique_3("abiola")==False)
```
