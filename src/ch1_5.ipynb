```python
def one_away(string1, string2):
  if abs(len(string1)-len(string2))>1:
    return False
  #case of replacement
  if len(string1)==len(string2):
    peer_unmatch_counts=0
    for index in range(len(string1)):
      if string1[index]!=string2[index]:
        if peer_unmatch_counts==1:
          return False 
        else:
          peer_unmatch_counts+=1
    return True

  #case of insertion/removal
  if abs(len(string1)-len(string2))==1:
    if len(string1)> len(string2):
      longest, shortest = string1, string2
    else:
      longest, shortest = string2, string1
    long_index, short_index = 0,0
    minimum_insertion_counts = 0
    for index in range(len(longest)):
      if longest[long_index] != shortest[short_index]:
        if minimum_insertion_counts==1:
          return False
        else:
          minimum_insertion_counts+=1
          long_index+=1
      else:
        if short_index== len(shortest)-1:
          return True
        else:
          long_index+=1
          short_index+=1
    return True
```

```python
assert(one_away("pale", "ple")==True)
assert(one_away("pales", "pale")==True)
assert(one_away("pale","bale")==True)
assert(one_away("pale","bake")==False)
```
