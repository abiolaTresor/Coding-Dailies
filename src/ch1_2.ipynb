```python
def check_permutation(string1, string2):
  if len(string1)==len(string2):
    quant_tuples = [(0,0)]*128
    for char in string1:
      #print(quant_tuples[ord(char)][0])
      quant_tuples[ord(char)]=(quant_tuples[ord(char)][0]+1,quant_tuples[ord(char)][1])
    for char in string2:
      quant_tuples[ord(char)]=(quant_tuples[ord(char)][0],quant_tuples[ord(char)][1]+1)
      if quant_tuples[ord(char)][1] > quant_tuples[ord(char)][0]:
        return False
    for tp in quant_tuples:
       if tp[0]<tp[1]:
         return False
    return True
  return False

    
```

```python
assert(check_permutation("","")==True)
assert(check_permutation("ab","ba")==True)
assert(check_permutation("abdefd","bafded")==True)
assert(check_permutation("abdefd","bafdedd")==False)
```

```python
#Simple and easily understandable
def check_permutation_2(string1, string2):
  if len(string1)==len(string2):
    sorted_string1 = "".join(sorted(string1))
    sorted_string2 = "".join(sorted(string2))
    return sorted_string1==sorted_string2
  return False
```

```python
assert(check_permutation_2("","")==True)
assert(check_permutation_2("ab","ba")==True)
assert(check_permutation_2("abdefd","bafded")==True)
assert(check_permutation_2("abdefd","bafdedd")==False)
```
